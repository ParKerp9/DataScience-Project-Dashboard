```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("videogamesales.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python

df.isnull().sum()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce', dayfirst=True)
```

```python
df.info()
```

```python
df['year'] = df['release_date'].dt.year
df = df.dropna(subset=['year'])
df['year'] = df['year'].astype(int)
```

```python
df[['release_date', 'year']].head()
```

```python
df = df.drop_duplicates() 
df['console'].unique()
df['genre'].unique() 
```

```python
df['decade'] = (df['year'] // 10) * 10
df = df.sort_values(by='total_sales', ascending=False)
```

```python
df.head()
```

```python
df.to_csv("cleaned_vgsales.csv", index=False)
```
